# **1. BINARY CLASSSIFICATION - PLAY TENNIS**

Cho tập dữ liệu huấn luyện mô hình Naive Bayes Classifier gồm các thuộc tính: "Outlook", "Temperature", "Humidity", "Wind"

<center width="100%"><img src="https://docs.google.com/uc?export=download&id=17HwzstJoMoWVdh-UCikbBZtfLkVzixGG" width="700px"></img></center>

**Cho sự kiện thử nghiệm:**

$\text{X = ("Outlook" = "Sunny", "Temperature" = "Cool", "Humidity" = "High", "Wind" = "Strong")}$

### **Câu hỏi 1:** Xác suất xảy ra sự kiện "Play Tennis"="Yes" và sự kiện "Play Tennis"="No" lần lượt là:

$\text{P("Play Tennis" = "Yes") = 6/10}$

$\text{P("Play Tennis" = "No") = 4/10}$

----> **A**

### **Câu hỏi 2:** Xác suất xảy ra sự kiện "Play Tennis"="Yes" khi sự kiện $\text{X}$ xảy ra là:

$\text{P("Play Tennis" = "Yes" | X)} ∝ \text{P("Sunny" | "Yes")} \times \text{P("Cool" | "Yes")} \times \text{P("High" | "Yes")} \times \text{P("Strong" | "Yes")} \times \text{P("Yes")}$

$\text{whereas:}$

$\displaystyle \text{P("Sunny" | "Yes")} = \frac{1}{6}$

$\displaystyle \text{P("Cool" | "Yes")} = \frac{3}{6}$

$\displaystyle \text{P("High" | "Yes")} = \frac{2}{6}$

$\displaystyle \text{P("Strong" | "Yes")} = \frac{1}{6}$

$\displaystyle \text{P("Yes")} = \frac{6}{10}$

$\displaystyle \Rightarrow \text{P("Play Tennis" = "Yes" | X)} ∝ \frac{1}{6} \times \frac{3}{6} \times \frac{2}{6} \times \frac{1}{6} \times \frac{6}{10} \approx 0.0028$

---> **B**

### **Câu hỏi 3:** Xác suất xảy ra sự kiện "Play Tennis"="No" khi sự kiện $\text{X}$ xảy ra là:

$\text{P("Play Tennis" = "No" | X)} ∝ \text{P("Sunny" | "No")} \times \text{P("Cool" | "No")} \times \text{P("High" | "No")} \times \text{P("Strong" | "No")} \times \text{P("No")}$

$\text{whereas:}$

$\displaystyle \text{P("Sunny" | "No")} = \frac{2}{4}$

$\displaystyle \text{P("Cool" | "No")} = \frac{1}{4}$

$\displaystyle \text{P("High" | "No")} = \frac{3}{4}$

$\displaystyle \text{P("Strong" | "No")} = \frac{2}{4}$

$\displaystyle \text{P("No")} = \frac{4}{10}$

$\displaystyle \Rightarrow \text{P("Play Tennis" = "No" | X)} ∝ \frac{2}{4} \times \frac{1}{4} \times \frac{3}{4} \times \frac{2}{4} \times \frac{4}{10} = 0.01875$

---> **C**

### **Câu hỏi 4:** Khi xảy ra sự kiện $\text{X}$, nhãn của "Play Tennis" sẽ là:

$\text{P("Play Tennis" = "No" | X)} > \text{P("Play Tennis" = "Yes" | X)}, \text{because } 0.01875 > 0.0028$

---> **B**

# **2. MULTI-LABEL CLASSIFICATION - TRAFFIC DATA**

Cho tập dữ liệu huấn luyện mô hình Naive Bayes Classifier gồm các thuộc tính: "Day", "Season", "Fog", "Rain"

<center width="100%"><img src="https://docs.google.com/uc?export=download&id=1Th1048PqUggNtB7f4jQ4yajA8jbeq-_9" width="650px"></img></center>

**Cho sự kiện thử nghiệm:**

$\text{X = ("Day" = "Weekday", "Season" = "Winter", "Fog" = "High", "Rain" = "Heavy")}$

### **Câu hỏi 5:** Xác suất xảy ra sự kiện "Class"="On Time", sự kiện "Class"="Late", sự kiện "Class"=" Very Late" và sự kiện "Class"="Cancelled" lần lượt là:

$\displaystyle \text{P("On Time")} = \frac{14}{20}$

$\displaystyle \text{P("Late")} = \frac{2}{20}$

$\displaystyle \text{P("Very Late")} = \frac{3}{20}$

$\displaystyle \text{P("Cancelled")} = \frac{1}{20}$

---> **A**

### **Câu hỏi 6:** Xác suất xảy ra sự kiện "Class"="On Time" khi sự kiện $\text{X}$ xảy ra là:

$\text{P("Class" = "On Time" | X)} ∝ \text{P("Weekday" | "On Time")} \times \text{P("Winter" | "On Time")} \times \text{P("High" | "On Time")} \times \text{P("Heavy" | "On Time")} \times \text{P("On Time")}$

$\text{whereas:}$

$\displaystyle \text{P("Weekday" | "On Time")} = \frac{9}{14}$

$\displaystyle \text{P("Winter" | "On Time")} = \frac{2}{14}$

$\displaystyle \text{P("High" | "On Time")} = \frac{4}{14}$

$\displaystyle \text{P("Heavy" | "On Time")} = \frac{2}{14}$

$\displaystyle \text{P("On Time")} = \frac{14}{20}$

$\displaystyle \Rightarrow \text{P("Play Tennis" = "On Time" | X)} ∝ \frac{9}{14} \times \frac{2}{14} \times \frac{4}{14} \times \frac{2}{14} \times \frac{14}{20} \approx 0.0026$

---> **C**

### **Câu hỏi 7:** Xác suất xảy ra sự kiện "Class"="Late" khi sự kiện $\text{X}$ xảy ra là:

$\text{P("Class" = "Late" | X)} ∝ \text{P("Weekday" | "Late")} \times \text{P("Winter" | "Late")} \times \text{P("High" | "Late")} \times \text{P("Heavy" | "Late")} \times \text{P("Late")}$

$\text{whereas:}$

$\displaystyle \text{P("Weekday" | "Late")} = \frac{1}{2}$

$\displaystyle \text{P("Winter" | "Late")} = \frac{2}{2}$

$\displaystyle \text{P("High" | "Late")} = \frac{1}{2}$

$\displaystyle \text{P("Heavy" | "Late")} = \frac{0}{2}$

$\displaystyle \text{P("Late")} = \frac{2}{20}$

$\displaystyle \Rightarrow \text{P("Play Tennis" = "Late" | X)} ∝ \frac{1}{2} \times \frac{2}{2} \times \frac{1}{2} \times \frac{0}{2} \times \frac{2}{20} = 0.0000$

---> **D**

### **Câu hỏi 8:** Xác suất xảy ra sự kiện "Class"= "Very Late" khi sự kiện $\text{X}$ xảy ra là:

$\text{P("Class" = "Very Late" | X)} ∝ \text{P("Weekday" | "Very Late")} \times \text{P("Winter" | "Very Late")} \times \text{P("High" | "Very Late")} \times \text{P("Heavy" | "Very Late")} \times \text{P("Very Late")}$

$\text{whereas:}$

$\displaystyle \text{P("Weekday" | "Very Late")} = \frac{3}{3}$

$\displaystyle \text{P("Winter" | "Very Late")} = \frac{2}{3}$

$\displaystyle \text{P("High" | "Very Late")} = \frac{1}{3}$

$\displaystyle \text{P("Heavy" | "Very Late")} = \frac{2}{3}$

$\displaystyle \text{P("Very Late")} = \frac{3}{20}$

$\displaystyle \Rightarrow \text{P("Play Tennis" = "Very Late" | X)} ∝ \frac{3}{3} \times \frac{2}{3} \times \frac{1}{3} \times \frac{2}{3} \times \frac{3}{20} \approx 0.0222$

---> **A**

### **Câu hỏi 9:** Xác suất xảy ra sự kiện "Class"= Cancelled" khi sự kiện $\text{X}$ xảy ra là:

$\text{P("Class" = "Cancelled" | X)} ∝ \text{P("Weekday" | "Cancelled")} \times \text{P("Winter" | "Cancelled")} \times \text{P("High" | "Cancelled")} \times \text{P("Heavy" | "Cancelled")} \times \text{P("Cancelled")}$

$\text{whereas:}$

$\displaystyle \text{P("Weekday" | "Cancelled")} = \frac{0}{1}$

$\displaystyle \text{P("Winter" | "Cancelled")} = \frac{0}{1}$

$\displaystyle \text{P("High" | "Cancelled")} = \frac{1}{1}$

$\displaystyle \text{P("Heavy" | "Cancelled")} = \frac{1}{1}$

$\displaystyle \text{P("Cancelled")} = \frac{1}{20}$

$\displaystyle \Rightarrow \text{P("Play Tennis" = "Cancelled" | X)} ∝ \frac{0}{1} \times \frac{0}{1} \times \frac{1}{1} \times \frac{1}{1} \times \frac{1}{20} = 0.0000$

---> **D**

### **Câu hỏi 10:** Dự đoán "Class" của sự kiện X là:

$\text{P("Play Tennis" = "Very Late" | X)} ∝ 0.0222$

---> **C**

$\text{P("Class" = "1" | X)} ∝ \text{P(X | "Class" = "1")} \times \text{P("Class" = "1")}$
